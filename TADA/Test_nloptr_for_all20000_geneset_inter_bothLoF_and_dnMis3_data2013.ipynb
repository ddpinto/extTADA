```python
%%R
library("data.table")

dirFile = "/usr/data/Documents/SaveDataFromServer/CombinedDataThreeDiseases/"
fileN = "result.nloptr.outerProduct.TADA2013.nonlinearNLOPT_GN_ISRES.both.all.LoF.Mis3.geneSet.all20000.txt"
nloptr.result <- fread(paste(dirFile, fileN, sep = ""))
```

```python
%%R
print(dim(nloptr.result))
```

```python
%%R -w 28 -h 56 -u cm
par(mfrow = c(9, 4))
for (ii in 1:dim(nloptr.result)[2]){
    hist(nloptr.result[[ii]], 100, main = ii)
    plot(density(nloptr.result[[ii]]), main = ii)
}
```

```python
%%R
round(apply(nloptr.result, 2, median), 3)
```

```python
%%R -w 28 -h 30 -u cm

dimK <- floor(dim(nloptr.result)[2]/2)

print(dimK)
par(mfrow = c(3, 2))
for (jj in 1:dimK)
plot(nloptr.result[[jj]], nloptr.result[[jj + dimK]], xlab = 'initial value', ylab = 'Value',
    main = round(cor(nloptr.result[[jj]], nloptr.result[[jj + dimK]], method = "spearman"), 2),
    ylim = range(nloptr.result[[jj]]))
```
