```python
%%R -w 28 -h 28 -u cm

full.cc.LoF.rho <- full.cc.LoF.rho[full.cc.LoF.rho[, 2]> 1, ]
par(mfrow = c(3, 2))
plot(full.cc.LoF.rho[, 4], full.cc.LoF.rho[, cLLK], xlab = "rho", ylab = 'logllk')
plot(full.cc.LoF.rho[, 1], full.cc.LoF.rho[, cLLK], xlab = "piX", ylab = 'logllk')

plot(full.cc.LoF.rho[, 2], full.cc.LoF.rho[, cLLK], xlab = "gamma.mean.cc", ylab = 'logllk')
plot(full.cc.LoF.rho[, 3], full.cc.LoF.rho[, cLLK], xlab = "beta.mean.cc", ylab = 'logllk')

plot(full.cc.LoF.rho[, 5], full.cc.LoF.rho[, cLLK], xlab = "nu0", ylab = 'logllk')

```

```python
%%R
par(mfrow = c(3, 2))
plot(full.cc.LoF.rho[, 4], full.cc.LoF.rho[, cLLK], xlab = "rho", ylab = 'logllk', ylim = c(lLimit, uLimit))

plot(full.cc.LoF.rho[, 1], full.cc.LoF.rho[, cLLK], xlab = "piX", ylab = 'logllk', ylim = c(lLimit, uLimit))

plot(full.cc.LoF.rho[, 2], full.cc.LoF.rho[, cLLK], xlab = "gamma.mean.cc", ylab = 'logllk', ylim = c(lLimit, uLimit))
plot(full.cc.LoF.rho[, 3], full.cc.LoF.rho[, cLLK], xlab = "beta.mean.cc", ylab = 'logllk', ylim = c(lLimit, uLimit))

plot(full.cc.LoF.rho[, 5], full.cc.LoF.rho[, cLLK], xlab = "nu0", ylab = 'logllk', ylim = c(lLimit, uLimit))
```

###Test only two parameters: pi and gamma.mean

```python
%%bash
ls /usr/data/Documents/SaveDataFromServer/GridResults/gridResults.2.dn*|grep -v "mis3" > list.2.dn.new.txt
#cat list.2.dn.new.txt
```

```python
%%R -w 28 -h 28 -u cm

list.2.dn.rdata <- readLines("list.2.dn.new.txt")
d.dn.grid <- NULL

for (ii in 1:length(list.2.dn.rdata)){
    load(list.2.dn.rdata[ii])
    t1 <- cbind(do.call(rbind, dn.grid), -as.numeric(llk))

    d.dn.grid <- rbind(d.dn.grid, t1)

    
}
```

```python
%%R
d.dn.grid <- d.dn.grid[d.dn.grid[, 3] > 0.1, ]
dim(d.dn.grid)
```

```python
%%R -w 28 -h 28 -u cm

par(mfrow = c(2, 2))
plot(d.dn.grid[, 1], d.dn.grid[, 4], xlab = 'pi', ylab = 'LogLikelihood')
plot(d.dn.grid[, 2], d.dn.grid[, 4], xlab = 'gamma.mean.dn.LoF', ylab = 'LogLikelihood')
abline(v = 20, col = 'red')
plot(d.dn.grid[, 3], d.dn.grid[, 4], xlab = 'beta.mean.dn.LoF', ylab = 'LogLikelihood')
```

```python
%%R -w 28 -h 20 -u cm
uLimit <- max(d.dn.grid[, 4]) + 1
lLimit <- uLimit - 10

par(mfrow = c(3, 1))
plot(d.dn.grid[, 1], d.dn.grid[, 4], xlab = 'pi', ylab = 'LogLikelihood', ylim = c(lLimit, uLimit))
plot(d.dn.grid[, 2], d.dn.grid[, 4], xlab = 'gamma.mean.dn.LoF', ylab = 'LogLikelihood', ylim = c(lLimit, uLimit))

plot(d.dn.grid[, 3], d.dn.grid[, 4], xlab = 'beta.mean.dn.LoF', ylab = 'LogLikelihood', ylim = c(lLimit, uLimit))
```

```python
%%R
d.dn.grid.highLLK <- d.dn.grid[d.dn.grid[, 4] > lLimit, ]

print(head(d.dn.grid.highLLK ))
print(tail(d.dn.grid.highLLK ))
print(summary(d.dn.grid.highLLK[, 2]))
message("\n\n")
print(summary(d.dn.grid.highLLK[, 1]))
message("\n\n")
print(summary(d.dn.grid.highLLK[, 3]))
```

```python
%%R -w 28 -h 28 -u cm
plot(d.dn.grid.highLLK[, 1], d.dn.grid.highLLK[, 2], lwd = d.dn.grid.highLLK[, 4] - lLimit, col = 'blue',
    xlab = 'pi', ylab = 'gamma.mean', cex = d.dn.grid.highLLK[, 4] - lLimit)

plot(d.dn.grid.highLLK[, 1], d.dn.grid.highLLK[, 3], lwd = d.dn.grid.highLLK[, 4] - lLimit, col = 'blue',
    xlab = 'pi', ylab = 'beta.mean', cex = d.dn.grid.highLLK[, 4] - lLimit)

plot(d.dn.grid.highLLK[, 2], d.dn.grid.highLLK[, 3], lwd = d.dn.grid.highLLK[, 4] - lLimit, col = 'blue',
    xlab = 'gamma.mean', ylab = 'beta.mean', cex = d.dn.grid.highLLK[, 4] - lLimit, ylim = c(0, 1))

abline(v = 20, col= 'red')
```

```python
%%R -w 28 -h 28 -u cm
library("plot3D")
scatter3D(z = d.dn.grid[, 4], x = d.dn.grid[, 1], y = d.dn.grid[, 2], xlab = 'pi', ylab = 'gammma.mean.LoF.dn',
         zlab = 'logLikelihood', theta = 10, d = 2,
         colkey = list(length = 0.5, width = 0.5, cex.clab = 0.75), ticktype = "detailed", 
          zlim = c(min(d.dn.grid[, 4]), uLimit + 200))
scatter3D(z = d.dn.grid[, 4], x = d.dn.grid[, 1], y = d.dn.grid[, 2], xlab = 'pi', ylab = 'gammma.mean.LoF.dn',
         zlab = 'logLikelihood', zlim = c(lLimit, uLimit + 50), ticktype = "detailed")
```

```python
%%R
scatter3D(z = d.dn.grid[, 4], x = d.dn.grid[, 1], y = d.dn.grid[, 2], xlab = 'pi', ylab = 'gammma.mean.LoF.dn',
         zlab = 'logLikelihood', phi = 0, bty = "g",  pch = 20, cex = 2, ticktype = "detailed")

scatter3D(z = d.dn.grid[, 4], x = d.dn.grid[, 1], y = d.dn.grid[, 2], xlab = 'pi', ylab = 'gammma.mean.LoF.dn',
         zlab = 'logLikelihood', phi = 0, bty = "g",  pch = 20, cex = 2, ticktype = "detailed", 
          zlim = c(lLimit, uLimit))
```

```python
%%R

x1 <- d.dn.grid[, 1]
x2 <- unique(x1)
y1 <- d.dn.grid[, 2]
y2 <- unique(y1)

z2 <- matrix(0, ncol = length(y2), nrow = length(x2))

for (ii in 1:length(x2)){
    for (jj in 1:length(y2)){
        z2[ii, jj] <- d.dn.grid[(d.dn.grid[, 1] == x2[ii]) & (d.dn.grid[, 2] == y2[jj]), 4]
    }}

```

```python
%%R
hist3D (z = z2, x = x2, y = y2,
         bty = "g", phi = 20,  theta = -60,
        xlab = 'pi', ylab = 'gammma.mean.LoF.dn',
         zlab = 'logLikelihood',
         col = "green", border = "black", shade = 0.8,
         ticktype = "detailed", space = 0.15, d = 2, cex.axis = 1e-9,
       
          zlim = c(lLimit, uLimit))
```
