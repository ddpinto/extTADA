```python
%%R
library("data.table")
##Only FMRP gene set
#nloptr.result <- fread("/usr/data/Documents/SaveDataFromServer/CombinedDataThreeDiseases/result.nloptr.outerProduct.geneSet.FMRP.txt")
#nloptr.result <- fread("/usr/data/Documents/SaveDataFromServer/CombinedDataThreeDiseases/result.nloptr.interProduct.geneSet.FMRP.txt")

dirFile = "/usr/data/Documents/SaveDataFromServer/CombinedDataThreeDiseases/"
#fileN = "result.nloptr.interProduct.nonlinearNLMMINB.bothLoF.dnMis3.geneSet.FMRP.new1.txt"

fileN <- "result.nloptr.outerProduct.nonlinearNLMMINB.bothLoF.dnMis3.geneSet.FMRP.new1.txt"
nloptr.result <- fread(paste(dirFile, fileN, sep = ""))
##All 20000 genes
#nloptr.result <- fread("/usr/data/Documents/SaveDataFromServer/CombinedDataThreeDiseases/result.nloptr.outerProduct.all20000genes.txt")
```

```python
%%R -w 28 -h 30 -u cm

par(mfrow = c(4, 3))
colNames <- c("pi", "gamma.mean.lof.dn", "beta.lof.dn", "gamma.mean.lof.cc", "beta.lof.cc",
             "rho1.lof", "nu1.lof", "rho0.lof", "nu0.lof",
              "gamma.mean.mis3.dn", "beta.lof.dn")
for (jj in 1:11)
plot(nloptr.result[[jj]], nloptr.result[[jj + 11]], 
     xlab = 'initial value', ylab = 'Value',
    main = paste(colNames[jj], "\n", 
                 round(cor(nloptr.result[[jj]], nloptr.result[[jj + 11]], method = "spearman"), 2), sep = ""),
    col = 'blue', xlim = range(nloptr.result[[jj]]), 
    ylim = range(nloptr.result[[jj]])    )
```
