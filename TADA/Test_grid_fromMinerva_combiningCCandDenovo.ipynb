```python
%%R
library("data.table")

data.cc.lof.both <- fread("/usr/data/Documents/SaveDataFromServer/GridResults/final.both.LoF.grid.txt")
```

```python
%%R
head(data.cc.lof.both)
```

```python
%%R
lof.cc.dn <- data.cc.lof.both

cLLK <- dim(lof.cc.dn)[2]
lof.cc.dn[[cLLK]] <- -lof.cc.dn[[cLLK]]
paraNames <- c("pi", "gamma.dn.mean", "beta.dn", "gamma.cc.mean", 
              "beta.cc", "rho1", "nu1")
head(lof.cc.dn)
```

```python
%%R -w 28 -h 28 -u cm

par(mfrow = c(4, 2))

for (jj in 1:7)
    plot(lof.cc.dn[[jj]] , lof.cc.dn[[8]], xlab = paraNames[jj], ylab = 'llk')

```

```python
%%R -w 28 -h 28 -u cm
maxLimit <- max(lof.cc.dn[[cLLK]]) + 2
minLimit <- maxLimit - 200


par(mfrow = c(4, 2))

for (jj in 1:7)
    plot(lof.cc.dn[[jj]], lof.cc.dn[[8]], xlab = paraNames[jj], ylab = 'llk', ylim = c(minLimit, maxLimit))

```
