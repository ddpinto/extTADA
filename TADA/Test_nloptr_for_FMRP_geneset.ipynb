```python
%%R
library("data.table")
##Only FMRP gene set
#nloptr.result <- fread("/usr/data/Documents/SaveDataFromServer/CombinedDataThreeDiseases/result.nloptr.outerProduct.geneSet.FMRP.txt")
#nloptr.result <- fread("/usr/data/Documents/SaveDataFromServer/CombinedDataThreeDiseases/result.nloptr.interProduct.geneSet.FMRP.txt")

nloptr.result <- fread("/usr/data/Documents/SaveDataFromServer/CombinedDataThreeDiseases/result.nloptr.interProduct.nonlinearNLOPT_GN_ISRES.bothLoF.dnMis3.geneSet.FMRP.txt")
##All 20000 genes
#nloptr.result <- fread("/usr/data/Documents/SaveDataFromServer/CombinedDataThreeDiseases/result.nloptr.outerProduct.all20000genes.txt")
```

```python
%%R
print(dim(nloptr.result))
```

```python
%%R -w 28 -h 56 -u cm
par(mfrow = c(12, 4))
for (ii in 1:dim(nloptr.result)[2]){
    hist(nloptr.result[[ii]], 100, main = ii)
    plot(density(nloptr.result[[ii]]), main = ii)
}
```

```python
%%R
round(apply(nloptr.result, 2, median), 3)
```

```python
%%R -w 28 -h 30 -u cm

par(mfrow = c(4, 3))
for (jj in 1:11)
plot(nloptr.result[[jj]], nloptr.result[[jj + 11]], xlab = 'initial value', ylab = 'Value',
    main = round(cor(nloptr.result[[jj]], nloptr.result[[jj + 11]], method = "spearman"), 2))
```
